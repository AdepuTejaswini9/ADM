```python
import matplotlib.pyplot as plt
import numpy as np
```

**Create Visualizations Using Matplotlib and Seaborn.**
1.   **Histogram**

```python
data = np.random.rand(1000)
# plt.hist(x=x,y=y,bins=30)
plt.hist(data, bins=30, color = 'red',edgecolor = 'black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

**2.**   **Bar chart**

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['sex'].value_counts().index, y=df['sex'].value_counts().values, palette='viridis')
plt.title('Bar Chart of Gender')
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
plt.show()
```

**3.**   **Box Plot**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['target'], y=df['age'], palette='coolwarm')
plt.title('Box Plot of Age by Heart Disease Target')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
```

**4. Correlation Heatmap**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**5. Violin Plot**

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='target', y='chol', data=df, palette='muted')
plt.title('Violin Plot of Cholesterol by Heart Disease Target')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Cholesterol')
plt.show()
```

**6. Count Plot**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='cp', data=df, palette='pastel')
plt.title('Count Plot of Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

**7.   Line chart**

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='age', y='thalach', data=df, marker='o', color='green')
plt.title('Line Chart of Age vs Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.show()
```
